 Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('Day_12_banking_data.csv')
```

```python
# Group by 'Account_Type' and calculate the total sum of 'Transaction_Amount'
account_type_sum = df.groupby('Account_Type')['Transaction_Amount'].sum()

# Plot the total sum of 'Transaction_Amount' per 'Account_Type' as a bar plot
account_type_sum.plot(kind='bar', color='skyblue', figsize=(10, 6))

# Adding labels and title
plt.title('Total Transaction Amount per Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the number of transactions per 'Branch'
branch_transaction_counts = df['Branch'].value_counts()

# Plot a pie chart for the percentage of transactions per 'Branch'
branch_transaction_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])

# Adding a title
plt.title('Percentage of Transactions per Branch')
plt.ylabel('')  # To remove the default 'Branch' label on the y-axis
plt.show()
```
