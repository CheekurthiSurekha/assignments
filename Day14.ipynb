Drug Effectiveness and Side Effects Comparison

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset from the CSV file
df = pd.read_csv('Day_13_Pharma_data.csv')  # Replace 'your_dataset.csv' with the actual file name

# Data Cleaning: Check for missing values and duplicates
print("Missing values in the dataset:\n", df.isnull().sum())
print("\nNumber of duplicate rows:", df.duplicated().sum())

# Visualization 1: Bar plot comparing average Effectiveness for each drug across different regions
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Effectiveness', hue='Product_Name', data=df)
plt.title('Average Effectiveness of Drugs by Region')
plt.xlabel('Region')
plt.ylabel('Average Effectiveness')
plt.show()

# Visualization 2: Violin plot to show the distribution of Effectiveness and Side_Effects for each product
plt.figure(figsize=(10, 6))
sns.violinplot(x='Product_Name', y='Effectiveness', data=df, inner="quart")
plt.title('Distribution of Effectiveness for Each Product')
plt.xlabel('Product Name')
plt.ylabel('Effectiveness')
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(x='Product_Name', y='Side_Effects', data=df, inner="quart")
plt.title('Distribution of Side Effects for Each Product')
plt.xlabel('Product Name')
plt.ylabel('Side Effects')
plt.show()

# Visualization 3: Pairplot to explore relationships between Effectiveness, Side_Effects, and Marketing_Spend
sns.pairplot(df[['Effectiveness', 'Side_Effects', 'Marketing_Spend']])
plt.suptitle('Pairplot: Effectiveness, Side Effects, and Marketing Spend', y=1.02)
plt.show()

# Visualization 4: Boxplot comparing Effectiveness for different trial periods
plt.figure(figsize=(8, 6))
sns.boxplot(x='Trial_Period', y='Effectiveness', data=df)
plt.title('Effectiveness Across Different Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Effectiveness')
plt.show()

# Visualization 5: Regression plot to analyze how Marketing_Spend affects drug Effectiveness
plt.figure(figsize=(8, 6))
sns.regplot(x='Marketing_Spend', y='Effectiveness', data=df, scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.title('Effectiveness vs Marketing Spend')
plt.xlabel('Marketing Spend')
plt.ylabel('Effectiveness')
plt.show()
```
